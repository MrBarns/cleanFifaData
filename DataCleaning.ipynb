```python
# Imports

import pandas as pd
import numpy as np
import re
import datetime as dt
```

```python
# Load dataframe

rawfifa = pd.read_csv(r"C:\Users\T. Barnes\Documents\Data Science\DataCleaningChallenge\Working_on\fifa21 raw data v2.csv")
rawfifa.head()
```

```python
# Modify column names

rawfifa.columns = [re.sub("[^a-zA-Z]", "_", x).strip().strip("_").lower() for x in rawfifa.columns]
rawfifa.columns
```

```python
# Remove newline character

rawfifa.club = [x.replace("\n", "") for x in rawfifa.club]
rawfifa.head()['club']
```

Success! Moving on

## contract, joined, loan_date_end

* check if the contract start year and joined year are the same.
* if the above is true, then the "joined" column is redundant

* also check if the contract end years and loan_date_end years are the same
* if true, then consider making loan_date_end a bool column instead

```python
# convert loan_date_end and "joined" cols to dates

rawfifa.loan_date_end = pd.to_datetime(rawfifa['loan_date_end'])
rawfifa.joined = pd.to_datetime(rawfifa.joined)
```

```python
rawfifa.joined
```

```python
# Split the contract column to get start and end years
contracts = [x.split("~") for x in rawfifa.contract]
contracts
```

On inspection of the resulting array, I can see that there are 3 different styles of values in the "contract" column:

* date range values
* "On Loan" values
* Free

Time for a closer look of the latter styles

```python
# View rows with "On Loan" values

rawfifa[rawfifa['contract'].str.contains("On Loan")][['contract', 'joined', 'loan_date_end',]]
```

This subset shows me that the "contract" column does not hold the start date for loan contracts. However, this date can be gotten from the "joined" column

```python
# View rows with "Free" values

rawfifa[rawfifa['contract'].str.contains("Free")][['contract', 'joined', 'loan_date_end', 'club']]
```

I have added the 'club' column to this subset and it confirms that this value represents those without a club contract.

This means the values in the "joined" column for these rows are irrelevant.

With this, I believe the best course of action is to:

* break up the "contract" col into "contract_start" and "contract_end" columns
* loan start and end years will be gotten from the "joined" and "loan_date_end" col respectively.
* all free values will be recorded as nans for both start and end
* drop joined col
* turn loan col to bool column
