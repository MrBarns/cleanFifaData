# Dates

Course of action decided upon from the previous file

* break up the "contract" col into "contract_start" and "contract_end" columns
* loan start and end years will be gotten from the "joined" and "loan_date_end" col respectively.
* all free values will be recorded as nans for both start and end
drop joined col
* turn loan col to bool column

Columns of interest are the
1. contract column
2. joined column
3. loan_date_end column

```python
# Imports

import numpy as np
import pandas as pd
import re
```

```python
# Load last checkpoint data

fifa = pd.read_csv(r"C:\Users\T. Barnes\Documents\Data Science\DataCleaningChallenge\Working_on\rawfifa_sun.csv")
fifa.head()
```

```python
# Make copy of loan dates for later

loan_copy = list(fifa.loan_date_end)
```

```python
fifa.joined = pd.to_datetime(fifa.joined)
fifa.loan_date_end = pd.to_datetime(fifa.loan_date_end)
```

```python
# Loop through columns for contract start and end years

contract_start = []
contract_end = []

for x, y, z in zip(fifa.contract, fifa.joined, fifa.loan_date_end):
    if "Free" in x:
        contract_start.append(float('nan'))
        contract_end.append(float('nan'))
    
    elif "Loan" in x:
        contract_start.append(y.year)
        contract_end.append(z.year)
        
    else:
        contract_start.append(x.split("~")[0].strip())
        contract_end.append(x.split("~")[1].strip())
```

```python
# Get booleans for is_on_loan field

is_on_loan = [False if type(x) == float else True for x in loan_copy]
```

I have chosen to preserve only the years here for consistency, since that is the one record that is available for all relevant fields. Other date values (month, date) are missing in several rows

```python
# Assign columns to dataframe

fifa['joined'] = contract_start
fifa['loan_date_end'] = contract_end
fifa['contract'] = is_on_loan
```

```python
# Rename columns

fifa.rename(columns={'contract': "is_loan_player",
                    "joined": "contract_start",
                    "loan_date_end": "contract_end"},
           inplace=True)
```

I decided to repurpose existing columns instead of dropping and adding fresh ones.
